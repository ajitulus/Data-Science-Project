```python
import pandas as pd
import numpy as np
```

```python
customers_data = pd.read_csv('/content/drive/MyDrive/Dataset/Data Science Project 1/customers_dataset.csv')
orders_items = pd.read_csv('/content/drive/MyDrive/Dataset/Data Science Project 1/order_items_dataset.csv')
orders_data = pd.read_csv('/content/drive/MyDrive/Dataset/Data Science Project 1/orders_dataset.csv')
products_data = pd.read_csv('/content/drive/MyDrive/Dataset/Data Science Project 1/products_dataset.csv')
```

### Challenge 1

```python
# gabung order data dan order item data
gabung_1 = orders_items.merge(orders_data, how='left', on='order_id')
gabung_1.head()
```

```python
# gabung_1 dengan product data
gabung_2 = gabung_1.merge(products_data, how='left', on='product_id')
gabung_2.head()
```

```python
#gabung_2 dengan customers data
gabung_data = gabung_2.merge(customers_data, how='left', on='customer_id')
gabung_data.head()
```

### Challenge 2

```python
datetime = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date']

for i in datetime:
 gabung_data[i] = pd.to_datetime(gabung_data[i])
```

```python
gabung_data.info()
```

### Challenge 3

```python
gabung_data.isnull().sum()
```

```python
type_null = pd.DataFrame(gabung_data.dtypes).T.rename(index = {0: 'Columns Type'})
type_null = type_null.append(pd.DataFrame(gabung_data.isnull().sum()).T.rename(index = {0: 'Amount of Null Values'}))
type_null = type_null.append(pd.DataFrame(gabung_data.isnull().sum()/gabung_data.shape[0]*100).T.rename(index = {0: 'Percentage of Null Values'}))
display(type_null)
```

### Challenge 4

```python
gabung_data['order_purchase_hour'] = gabung_data['order_purchase_timestamp'].dt.hour
gabung_data['order_purchase_day_name'] = gabung_data['order_purchase_timestamp'].dt.day_name()
gabung_data['order_purchase_weekday'] = gabung_data['order_purchase_timestamp'].dt.day_name()
```

```python
gabung_data['order_delivered_customer_hour'] = gabung_data['order_delivered_customer_date'].dt.hour
gabung_data['order_delivered_customer_day_name'] = gabung_data['order_delivered_customer_date'].dt.day_name()
gabung_data['order_delivered_customer_weekday'] = gabung_data['order_delivered_customer_date'].dt.day_name()
```

```python
gabung_data.head()
```

### Challenge 5

```python
c5 = pd.pivot_table(gabung_data, values='customer_id', index=['customer_city'], aggfunc=pd.Series.nunique)
c5.head()
```

### Challange 6

```python
c6 = c5.reset_index()
c6.sort_values(by='customer_id', ascending=False)[:5]
```

### Challenge 7

```python
top_5 = gabung_data[(gabung_data['customer_city'] == 'sao paulo') | (gabung_data['customer_city'] == 'rio de janeiro') |
                                      (gabung_data['customer_city'] == 'belo horizonte') | (gabung_data['customer_city'] == 'brasilia') |
                                      (gabung_data['customer_city'] == 'curitiba')]

top_5.head()
```

```python
top_5.groupby('customer_city').agg(average_hour = ('order_purchase_hour','mean'),
                                               median_hour = ('order_purchase_hour', 'median'))
```

### Challenge 8 

```python
c8 = pd.pivot_table(top_5, values='customer_id', index=['customer_city'], columns=['order_purchase_weekday'], aggfunc = pd.Series.nunique)
c8 = c8.reset_index()
c8 = c8.rename_axis(None, axis=1).reset_index(drop=True)
c8
```

### Challenge 9

```python
c8.style.background_gradient(cmap='Greens', axis=1)
```
